# Exploring NLTK

```python
# Open the text file
text_file = open("text.txt")
```

```python
# Read the text
text = text_file.read()
```

```python
# data type of the text
print(type(text))
```

```python
# text
print(text)
```

```python
# Length of the text
print(len(text)) # characters in the text file
```

```python
# Import required libraries:
import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
```

**Sentence Tokenizing**

```python
# Tokenize the text by sentences:
sentences = sent_tokenize(text)
sentences
```

```python
# Length of the sentence
print(len(sentences))
```

**Word Tokenization**

```python
# Tokenize the text by words:
words = word_tokenize(text)
words
```

```python
# Length of the words
print(len(words))
```

**Frequency distribution**

```python
# Importing necessary libraries:
from nltk.probability import FreqDist

# Find the frequency distribution 
fdist = FreqDist(words)

# Print 10 most common words
fdist.most_common(20)
```

```python
# Plot the graph for fdist:
import matplotlib.pyplot as plt

fdist.plot(10)
```

**Remove punctuation marks**

```python
# Empty words list to store words
word_no_punc = []

# Removing punctuation marks
for w in words:
    if w.isalpha():
        word_no_punc.append(w.lower())

print(len(word_no_punc))
```

```python
# Plotting graph without punctuation marks
fdist = FreqDist(word_no_punc)
fdist.most_common(10)
```

```python
fdist.plot(10)
```

```python
from nltk.corpus import stopwords
stopwords = stopwords.words("english")

# Removing stopwords
clean_words = []

for w in word_no_punc:
    if w not in stopwords:
        clean_words.append(w)

print(clean_words)
```

```python
# Plotting graph without punctuation marks

fdist = FreqDist(clean_words)
fdist.most_common(10)
```

```python
# Frequency distribution
fdist.plot(10)
```
